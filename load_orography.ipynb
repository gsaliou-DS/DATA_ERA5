```python
import netCDF4 as nc
import numpy as np
import matplotlib.pyplot as plt

def load_orography(file_path):
    # Open the NetCDF file
    ds = nc.Dataset(file_path, 'r')

    # Read the orography data
    orography = ds.variables['z'][:]  # Assuming 'orog' is the variable name for orography
    print(orography.shape)
    
    # Read the latitude and longitude data
    latitudes = ds.variables['latitude'][:]
    longitudes = ds.variables['longitude'][:]
    
    # Close the dataset
    ds.close()
    
    return orography, latitudes, longitudes

def plot_orography(orography, latitudes, longitudes):
    plt.figure(figsize=(10, 5))
    plt.contourf(longitudes, latitudes, orography, cmap='terrain')
    plt.colorbar(label='Orography (meters)')
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    plt.title('Global Orography')
    plt.show()

if __name__ == '__main__':
    # Path to the NetCDF file containing orography data
    #file_path = 'geopotential_0.25_local.nc'
    file_path = 'geopotential_2.5_global.nc'
    
    # Load the orography data
    orography, latitudes, longitudes = load_orography(file_path)
    g = 9.80665
    #orography = orography.reshape(81, 121) / g
    orography = orography.reshape(73, 144) / g
    print(orography.shape, latitudes.shape, longitudes.shape)
    print(orography)
    print(latitudes)
    print(longitudes)

    # Plot the orography data
    plot_orography(orography, latitudes, longitudes)
```
